# Open AI Gym exploration

```python
import gym
import gym_2048
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
env = gym.make('2048-v0')
```

## Random agent

```python
# random player
def random_agent(n):
    env.reset()
    for _ in range(n):
        env.step(env.action_space.sample()) # take a random action
    env.render()
    env.close()
```

```python
#Average score for 1000 parties played by random agent with 1000 moves
score_list = []
max_list = []
for i in range(1000):
    random_agent(1000)
    score_list.append(env.__dict__['score'])
    max_list.append(np.amax(env.__dict__['Matrix']));
```

```python
max_values = sorted(list(set(max_list)))
max_dict = {}
for max_case in max_values:
    max_dict[max_case] = max_list.count(max_case)

random_max_cell_df = pd.DataFrame(max_dict, index = [0]).T.reset_index()
random_max_cell_df.columns = ['Max_case','random']
random_max_cell_df
```

```python
fig = plt.figure(figsize = (20,5))
ax1 = fig.add_subplot(1,2,1)
ax1.hist(score_list)
ax1.set_title('score distribution')
ax2 = fig.add_subplot(1,2,2)
ax2.hist(max_list)
ax2.set_title('max distribution');
```

## Stupid agent

```python
def stupid_agent(n):
    env.reset()
    for i in range(50):
        for _ in range(n):
            env.step(0)
            env.step(1)
            env.step(0)
            env.step(1)
            env.step(0)
            env.step(1)
        env.step(2)
    env.render()
    env.close()
```

```python
stupid_score_list = []
stupid_max_list = []
for i in range(1000):
    stupid_agent(10)
    stupid_score_list.append(env.__dict__['score'])
    stupid_max_list.append(np.amax(env.__dict__['Matrix']));
```

```python
fig = plt.figure(figsize = (20,10))
ax1 = fig.add_subplot(2,2,1)
ax1.hist(stupid_score_list, color = 'r')
ax1.set_title('stupid score distribution')
ax1.set_xlim((0,7000))
ax2 = fig.add_subplot(2,2,2)
ax2.hist(stupid_max_list, color = 'r')
ax2.set_title('stupid max distribution')
ax2.set_xlim((0,600))
ax3 = fig.add_subplot(2,2,3)
ax3.hist(score_list, color = 'b')
ax3.set_title('random score distribution')
ax3.set_xlim((0,7000))
ax4 = fig.add_subplot(2,2,4)
ax4.hist(max_list, color = 'b')
ax4.set_title('random max distribution')
ax4.set_xlim((0,600));
```

```python
plt.hist(stupid_max_list, color = 'r', label = 'stupid')
plt.hist(max_list, color = 'b', label = 'random')
plt.legend();
```

```python
stupid_max_values = sorted(list(set(stupid_max_list)))
stupid_max_dict = {}
for max_case in stupid_max_values:
    stupid_max_dict[max_case] = stupid_max_list.count(max_case)

stupid_max_cell_df = pd.DataFrame(stupid_max_dict, index = [0]).T.reset_index()
stupid_max_cell_df.columns = ['Max_case','stupid']
stupid_max_cell_df
```

```python
games_df = stupid_max_cell_df.merge(random_max_cell_df, on = 'Max_case', how = 'outer').fillna(0).set_index('Max_case')
games_df.random = games_df.random.astype('int')
games_df.random = 100*games_df.random/games_df.random.sum()
games_df.stupid = 100*games_df.stupid/games_df.stupid.sum()
games_df['% variation'] = games_df.stupid - games_df.random
games_df
```
